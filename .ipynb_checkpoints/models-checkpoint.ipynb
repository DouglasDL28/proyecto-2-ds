```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("./data/clean_data.csv")
```

```python
data.head()
```

#### Feature selection

```python
data.columns
```

```python
output = 'default_payment_next_month'

cols = [ f for f in data.columns if data.dtypes[ f ] != "object"]
cols.remove(output)
cols.remove('id')
cols.remove('Unnamed: 0')
```

```python
features = cols
```

```python
features.remove('bill_amt1')
features.remove('bill_amt2')
features.remove('bill_amt3')
features.remove('bill_amt4')
features.remove('bill_amt5')
features.remove('bill_amt6')
features.remove('marriage_3')
features.remove('education_4')
features.remove('education_5')
features.remove('education_6')
```

##### Data scaling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X = data[features].values    
y = data[output].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scX = StandardScaler()
X_train = scX.fit_transform(X_train)
X_test = scX.transform(X_test)
```

## Linear Regression Model

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, mean_squared_error
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import normalize, MinMaxScaler
from scikitplot.metrics import plot_roc
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
model1 = LogisticRegression(solver='liblinear', C=10)
model1.fit(X_train, y_train)

y_pred = model1.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print('\nAccuracy Score for linear regression model: ', accuracy_score(y_test,y_pred))
```

```python
plot_roc(y_test, model1.predict_proba(X_test))
plt.show()
```

## SVM Model

```python
from sklearn import svm
from sklearn.metrics import accuracy_score, recall_score, precision_score

svm_classifier = svm.SVC(
    probability=True,
    kernel="rbf"
)
svm_classifier.fit(X_train, y_train)
y_pred = svm_classifier.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix for model")
print(cm)
```

```python
results = {}
results['accuracy']  = accuracy_score(y_test, y_pred)
results['recall']    = recall_score(y_test, y_pred, average=None)
results['precision'] = precision_score(y_test, y_pred, average=None)

for key, value in results.items():
    print("%s: %s" % (key, value))
```

```python
plot_roc(y_test, svm_classifier.predict_proba(X_test))
plt.show()
```

```python
# minimizando los falsos positivos
svm_classifier = svm.SVC(
    C=0.1, 
    kernel="rbf", 
    class_weight={0: 1, 1: 2},
    probability=True
)
svm_classifier.fit(X_train, y_train)
y_pred = svm_classifier.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix for model")
print(cm)
```

```python
results['accuracy']  = accuracy_score(y_test, y_pred)
results['recall']    = recall_score(y_test, y_pred, average=None)
results['precision'] = precision_score(y_test, y_pred, average=None)

for key, value in results.items():
    print("%s: %s" % (key, value))
```

```python
plot_roc(y_test, svm_classifier.predict_proba(X_test))
plt.show()
```

## Random Forest Model

```python
rf_model = RandomForestClassifier(random_state=0)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print('\nAccuracy Score for Random Forest model: ', accuracy_score(y_test,y_pred))
```

```python
plot_roc(y_test, rf_model.predict_proba(X_test))
plt.show()
```
